```python
jogos["ano"] = jogos["data"].apply(lambda date: int(date[-4:]))
jogos = jogos.sort_values(["ano"])
# jogos = jogos.set_index("ano")
jogos
```

```python
series_years = jogos["ano"]
# print(series_years)
for year in series_years.unique():
	print(year, ":", len(jogos[series_years == year]))
```

```python
class ChampionshipAnalyser():

	def __init__(self, ano_referencia):

		self.df_campeonato = jogos[series_years == ano_referencia]
		self.times = self.df_campeonato["mandante"].unique()
		self.tabela_classificacao = self._get_tabela_classificacao()
		self.campeao = self.tabela_classificacao.iloc[0]["time"]

	def _get_tabela_classificacao(self):
		dict_classificacao = {}
		for time in self.times:
			jogos_como_mandante = self.df_campeonato[self.df_campeonato["mandante"] == time]
			jogos_como_visitante = self.df_campeonato[self.df_campeonato["visitante"] == time]

			total_jogos = len(jogos_como_mandante) + len(jogos_como_visitante)

			# vitorias
			vitorias_como_mandante = jogos_como_mandante[jogos_como_mandante["mandante_Placar"] > jogos_como_mandante["visitante_Placar"]]
			vitorias_como_visitante = jogos_como_visitante[jogos_como_visitante["mandante_Placar"] < jogos_como_visitante["visitante_Placar"]]

			total_vitorias = len(vitorias_como_mandante) + len(vitorias_como_visitante)
			lista_vitorias = dict_classificacao.get("vitorias", [])
			lista_vitorias.append(total_vitorias)
			dict_classificacao["vitorias"] = lista_vitorias

			# derrotas
			derrotas_como_mandante = jogos_como_mandante[jogos_como_mandante["mandante_Placar"] < jogos_como_mandante["visitante_Placar"]]
			derrotas_como_visitante = jogos_como_visitante[jogos_como_visitante["mandante_Placar"] > jogos_como_visitante["visitante_Placar"]]

			total_derrotas = len(derrotas_como_mandante) + len(derrotas_como_visitante)
			lista_derrotas = dict_classificacao.get("derrotas", [])
			lista_derrotas.append(total_derrotas)
			dict_classificacao["derrotas"] = lista_derrotas

			# empates
			total_empates = total_jogos - (total_vitorias + total_derrotas)
			lista_empates = dict_classificacao.get("empates", [])
			lista_empates.append(total_empates)
			dict_classificacao["empates"] = lista_empates

		dict_classificacao["time"] = self.times

		tabela_classificacao = pd.DataFrame(dict_classificacao)
		tabela_classificacao["pontos"] = tabela_classificacao.apply(lambda row: 3 * row["vitorias"] + row["empates"], axis=1)
		tabela_classificacao = tabela_classificacao.sort_values(["pontos"], ascending=False)

		return tabela_classificacao

	def get_champion_games(self):
		jogos_campeao = self.df_campeonato[(self.df_campeonato["mandante"] == self.campeao) | (self.df_campeonato["visitante"] == self.campeao)]
		return jogos_campeao
	
	def get_coachs_champion(self):
		champion_games = self.get_champion_games()
		jogos_como_mandante = champion_games[champion_games["mandante"] == self.campeao]
		jogos_como_visitante = champion_games[champion_games["visitante"] == self.campeao]

		coachs_set_1 = set(jogos_como_mandante["tecnico_mandante"].unique())
		coachs_set_2 = set(jogos_como_visitante["tecnico_visitante"].unique())
		
		return coachs_set_1.union(coachs_set_2)
```

```python
analyser = ChampionshipAnalyser(2010)

print(analyser.tabela_classificacao.head())

analyser.get_champion_games().head()
```

```python
for year in range(2003, 2023):
	analyser = ChampionshipAnalyser(year)
	champion_games = analyser.get_champion_games()
	print(year, analyser.campeao, analyser.get_coachs_champion())
```
