```python
jogos["ano"] = jogos["data"].apply(lambda date: int(date[-4:]))
jogos = jogos.sort_values(["ano"])
# jogos = jogos.set_index("ano")
jogos
```

```python
series_years = jogos["ano"]
# print(series_years)
for year in series_years.unique():
	print(year, ":", len(jogos[series_years == year]))
```

```python
df_campeonato_2006 = jogos[series_years == 2006]
times = df_campeonato_2006["mandante"].unique()
len(times)
```

```python
dict_classificacao = {}
for time in times:
	jogos_como_mandante = df_campeonato_2006[df_campeonato_2006["mandante"] == time]
	jogos_como_visitante = df_campeonato_2006[df_campeonato_2006["visitante"] == time]

	total_jogos = len(jogos_como_mandante) + len(jogos_como_visitante)

	# vitorias
	vitorias_como_mandante = jogos_como_mandante[jogos_como_mandante["mandante_Placar"] > jogos_como_mandante["visitante_Placar"]]
	vitorias_como_visitante = jogos_como_visitante[jogos_como_visitante["mandante_Placar"] < jogos_como_visitante["visitante_Placar"]]

	total_vitorias = len(vitorias_como_mandante) + len(vitorias_como_visitante)
	lista_vitorias = dict_classificacao.get("vitorias", [])
	lista_vitorias.append(total_vitorias)
	dict_classificacao["vitorias"] = lista_vitorias

	# derrotas
	derrotas_como_mandante = jogos_como_mandante[jogos_como_mandante["mandante_Placar"] < jogos_como_mandante["visitante_Placar"]]
	derrotas_como_visitante = jogos_como_visitante[jogos_como_visitante["mandante_Placar"] > jogos_como_visitante["visitante_Placar"]]

	total_derrotas = len(derrotas_como_mandante) + len(derrotas_como_visitante)
	lista_derrotas = dict_classificacao.get("derrotas", [])
	lista_derrotas.append(total_derrotas)
	dict_classificacao["derrotas"] = lista_derrotas

	# empates
	total_empates = total_jogos - (total_vitorias + total_derrotas)
	lista_empates = dict_classificacao.get("empates", [])
	lista_empates.append(total_empates)
	dict_classificacao["empates"] = lista_empates

dict_classificacao["time"] = times
```

```python
tabela_classificacao = pd.DataFrame(dict_classificacao)
tabela_classificacao["pontos"] = tabela_classificacao.apply(lambda row: 3 * row["vitorias"] + row["empates"], axis=1)
tabela_classificacao.sort_values(["pontos"], ascending=False)
```
